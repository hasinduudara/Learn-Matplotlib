```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([10, 20, 15, 5])

line_styles = dict(marker=".",
                   markersize=30,
                   markerfacecolor="black",
                   markeredgecolor="black",
                   linestyle="solid",
                   linewidth=4,
                   color="gray")

plt.plot(x, y1, **line_styles)
plt.plot(x, y2, **line_styles)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([10, 20, 25, 5])
y3 = np.array([5, 10, 15, 20])

line_styles = dict(marker=".",
                   markersize=30,
                   markerfacecolor="black",
                   markeredgecolor="black",
                   linestyle="solid",
                   linewidth=4,)

plt.plot(x, y1, color="gray", **line_styles)
plt.plot(x, y2, color="red", **line_styles)
plt.plot(x, y3, color="green", **line_styles)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 25])
y2 = np.array([10, 20, 25, 5])
y3 = np.array([5, 10, 15, 20])

plt.title("Time Chart", fontsize=20,
                        family="Arial",
                        fontweight="bold",
                        color="blue")

plt.xlabel("Year", fontsize=15,
                   family="Arial",
                   fontweight="bold",
                   color="red")

plt.ylabel("Student", fontsize=15,
                      family="Arial",
                     fontweight="bold",
                     color="red")

plt.tick_params(axis="both",
                colors="green")

plt.plot(x, y1, color="gray")
plt.plot(x, y2, color="red")
plt.plot(x, y3, color="green")

plt.xticks(x)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

languages = ["Maths", "Science", "ICT"]
marks = np.array([85, 65, 30])
color = ["green", "red", "yellow"]

plt.pie(marks, labels=languages,
               autopct="%1.1f%%",
               colors=color,
               explode=[0, 0, 0.1],
               shadow=True,
               startangle=90)

plt.title("College Data")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8]) # Hours studied
y1 = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87]) # Grades

x2 = np.array([1, 2, 3, 3, 4, 5, 6, 6, 7, 8, 9]) # Hours studied
y2 = np.array([23, 45, 50, 34, 70, 65, 73, 80, 57, 68, 70]) # Grades

plt.scatter(x1, y1, color="red",
                  alpha=0.8,
                  s=100,
                  label="Class A")

plt.scatter(x2, y2, color="green",
                  alpha=0.8,
                  s=100,
                  label="Class A")

plt.title("Test Scores")
plt.xlabel("Hours studied")
plt.ylabel("Grades")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10,
                 color="red",
                 edgecolor="black")

plt.title("Exam Scores")
plt.xlabel("Scores")
plt.ylabel("# of students")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])

figure, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, x*2, color="red")
axes[0, 0].set_title("x*2")

axes[0, 1].plot(x, x**2, color="green")
axes[0, 1].set_title("x**2")

axes[1, 0].plot(x, x**3, color="yellow")
axes[1, 0].set_title("x**3")

axes[1, 1].plot(x, x**4, color="black")
axes[1, 1].set_title("x**4")

plt.tight_layout()
plt.show()
```
