```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])

plt.plot(x, y, marker=".",
               markersize=30,
               markerfacecolor="black",
               markeredgecolor="black",
               linestyle="solid",
               linewidth=4,
               color="gray")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([10, 20, 15, 5])

line_styles = dict(marker=".",
                   markersize=30,
                   markerfacecolor="black",
                   markeredgecolor="black",
                   linestyle="solid",
                   linewidth=4,
                   color="gray")

plt.plot(x, y1, **line_styles)
plt.plot(x, y2, **line_styles)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([10, 20, 25, 5])
y3 = np.array([5, 10, 15, 20])

line_styles = dict(marker=".",
                   markersize=30,
                   markerfacecolor="black",
                   markeredgecolor="black",
                   linestyle="solid",
                   linewidth=4,)

plt.plot(x, y1, color="gray", **line_styles)
plt.plot(x, y2, color="red", **line_styles)
plt.plot(x, y3, color="green", **line_styles)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 25])
y2 = np.array([10, 20, 25, 5])
y3 = np.array([5, 10, 15, 20])

plt.title("Time Chart", fontsize=20,
                        family="Arial",
                        fontweight="bold",
                        color="blue")

plt.xlabel("Year", fontsize=15,
                   family="Arial",
                   fontweight="bold",
                   color="red")

plt.ylabel("Student", fontsize=15,
                      family="Arial",
                     fontweight="bold",
                     color="red")

plt.tick_params(axis="both",
                colors="green")

plt.plot(x, y1, color="gray")
plt.plot(x, y2, color="red")
plt.plot(x, y3, color="green")

plt.xticks(x)

plt.show()
```
