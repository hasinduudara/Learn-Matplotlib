```python
import matplotlib.pyplot as plt

x = [2023, 2024, 2025, 2026]
y = [15, 25, 30, 20]

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])

plt.plot(x, y, marker=".",
               markersize=30,
               markerfacecolor="black",
               markeredgecolor="black",
               linestyle="solid",
               linewidth=4,
               color="gray")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([10, 20, 15, 5])

line_styles = dict(marker=".",
                   markersize=30,
                   markerfacecolor="black",
                   markeredgecolor="black",
                   linestyle="solid",
                   linewidth=4,
                   color="gray")

plt.plot(x, y1, **line_styles)
plt.plot(x, y2, **line_styles)
plt.show()
```
